```python
#code cell for running BFS

from collections import deque
import numpy as np
def bfs ():
    #define start and goal nodes
    start = np.array([[8 , 6 , 7],
                  [2 , 5 , 4],
                  [3 , 0 , 1]])
    goal = np.array([[1 , 2 , 3],
                     [4  , 5 ,6],
                     [7 , 8 , 0]])
    #Defin required variables for further creating text files
    start_tup = tuple(start.flatten())
    index = 0
    index_up = 0
    dic = {start_tup : index}
    explored_states = [start]
    nodes_info = [[0 , -1 , start_tup]]
    visited = set()
    #Define the queue
    queue = deque([(start, [start])])
    reached = 0
    p_node_index = 0
    node_index = 0
    no_solution = 0
    #start the algo
    while reached == 0 :
        if (len(queue) == 0) :
            no_solution = 1
            print('no solution')
            break
        current_state , path = queue.popleft()
        moves = []
        zero_position = np.argwhere(current_state == 0)[0]
        row, col = zero_position
        possible_moves = [(row , col-1), (row + 1, col), (row, col + 1), (row - 1, col)]
        # Carry out necessary checks
        if tuple(current_state.flatten()) not in visited:
                current_state_tup = tuple(current_state.flatten())
                visited.add(current_state_tup)
                p_index = dic[current_state_tup]
                for move in possible_moves:
                    new_row, new_col = move
                    if 0 <= new_row < 3 and 0 <= new_col < 3:
                        index_up += 1
                        new_state = current_state.copy()
                        new_state[row, col], new_state[new_row, new_col] = new_state[new_row, new_col], new_state[row, col]
                        moves.append(new_state)
                        explored_states.append(new_state)
                        nodes_info.append([index_up , p_index , tuple(new_state.flatten())])
                        dic[tuple(new_state.flatten())] = index_up
                for move in moves:
                    queue.append((move, path + [move]))
                    if (np.array_equal(move , goal)):
                        print("Goal state reached")
                        path.append(goal)
                        reached = 1
                        break
    return path , nodes_info , no_solution, explored_states               

path , nodes_info, no_solution, explored_states = bfs()
if (no_solution == 1):
    print('cannot print path')
else:
    for step in path:
        print(step)
```
